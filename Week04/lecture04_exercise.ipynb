```python
import pandas as pd
import plotly.express as px


# Dataset: Gapminder — GDP, Life Expectancy, Population by Country
# Source: Gapminder Foundation (gapminder.org)

df = pd.read_csv("./gapminder.csv")

print(f"Loaded: {len(df)} rows")
print(df.head())
```

```python
# explore

print(df.info())
print("\nContinents:", df['Continent'].unique())
print("\nSummary Statistics")
print(df.describe().round(1))
```

## **Task 1 — Scatter: life expectancy change over time**

```python
# Choose continent
continent_choice = "Europe"

# Filter continent
task1_df = df[df["Continent"] == continent_choice]

# Create scatter plot
fig1 = px.scatter(
    task1_df,

    x="GDP_per_capita",
    y="Life_expectancy",

    hover_name="Country",

    color="Continent",

    log_x=True,

    title=f"{continent_choice}: GDP vs Life Expectancy (2000 → 2007)",
    labels={
        "GDP_per_capita": "GDP per Capita (Log Scale)",
        "Life_expectancy": "Life Expectancy"
    }
)

# Styling
fig1.update_traces(
    marker=dict(
        size=12,
        opacity=0.8
    )
)

fig1.update_layout(
    template="simple_white",
    title_font_size=20,
    showlegend=False
)

# Annotation
fig1.add_annotation(
    x=20000,
    y=82,
    text="Higher-income countries generally<br>show higher life expectancy.",
    showarrow=True,
    arrowhead=2
)

fig1.show()
```

## **Task 2 — Bubble chart: tell a story**

```python
# Choose continent to highlight
highlight_continent = "Asia"

# Create highlight column
df["highlight"] = df["Continent"].apply(
    lambda x: highlight_continent if x == highlight_continent else "Other Continents"
)

# Create bubble chart
fig2 = px.scatter(
    df,

    x="GDP_per_capita",
    y="Life_expectancy",

    size="Population",

    color="highlight",

    hover_name="Country",

    log_x=True,

    size_max=60,
    title=f"2007 Global View: {highlight_continent} vs Rest of the World",
    labels={
        "GDP_per_capita": "GDP per Capita (Log Scale)",
        "Life_expectancy": "Life Expectancy",
        "Population": "Population"
    },

    color_discrete_map={
        highlight_continent: "#d62728",
        "Other Continents": "#c7c7c7"
    }
)

# Styling
fig2.update_traces(
    marker=dict(
        opacity=0.75,
        line=dict(width=0.5, color="white")
    )
)

fig2.update_layout(
    template="simple_white",
    title_font_size=20,
    showlegend=False
)

# Annotation
fig2.add_annotation(
    x=4000,
    y=72,
    text="Asia includes both low-income and high-income economies,<br>creating a wide GDP spread.",
    showarrow=True,
    arrowhead=2
)

fig2.show()
```
